```python
from datetime import datetime

import pandas as pd

# Открываю полученные таблички, первая из трех
clin_res=pd.read_csv('clin_res.csv', encoding = 'cp1251')
# Удаляю старый индекс
del clin_res['Unnamed: 0']
clin_res.shape
```

```python
from datetime import datetime
print (clin_res.head())


print(min(clin_res['date']), max(clin_res['date']))
# Колонки с датами судя по всему надо отформатировать в даты

date_2 = pd.to_datetime(clin_res.date, format='%d.%m.%Y')

print(len(date_2))
print(min(date_2), max(date_2))
    
```

```python
# Делаю датафрейм с колонкой отформатированных дат

nw_tb=clin_res.join(date_2, how='outer', lsuffix='1')
nw_tb.shape
```

```python
# Открываю полученные таблички, вторая
clin_res2=pd.read_csv('clin_res2.csv', encoding = 'cp1251')
# Опять удаляю старый индекс
del clin_res2['Unnamed: 0']
# Добавляю две колонки с началом и концом исследований, отформатированными в виде дат
date_3 = pd.to_datetime(clin_res2['Конец иссл.'], format='%d.%m.%Y')
date_4 = pd.to_datetime(clin_res2['Начало иссл.'], format='%d.%m.%Y')
print(clin_res2.shape)
print(len(date_3), len(date_4))
clin_res2=clin_res2.join(date_3,how='left', lsuffix='1')
clin_res2=clin_res2.join(date_4,how='left', lsuffix='1')
print(clin_res2.shape)
clin_res2.head()

#clin_res2.loc['543']
```

```python
# Собирается датафрейм, из собранных данных
grab_table=clin_res2.join(nw_tb)
#grab_table.set_index('nom_rki',inplace=True)
grab_table.shape
```

```python
#Добавляется колонка, с количеством организаций, и организациями, 
#для этого из колонки "Организации" делаем срезы строк с разделителем по запятой с пробелом (для второй колонки),
#и подсчитываем их длину (для первой колонки)
org_lst= grab_table['Организации'].str.split(', ') 

org_col=[len(x) for x in org_lst]

grab_table['Organizacyi']=org_lst
grab_table ['Orgs'] = org_col
print(grab_table.shape)
grab_table.head()
```

```python
# Открываю полученные таблички, третья (источник: открытые данные, Минздрав)
clin_res_mz=pd.read_csv("data-20150930-structure.csv", encoding = 'cp1251')
# Здесь надо отформатировать в дата три колонки
date_5 = pd.to_datetime(clin_res_mz['Начало (дата)'], format='%d.%m.%Y')
date_6 = pd.to_datetime(clin_res_mz['Окончание (дата)'], format='%d.%m.%Y')
date_7 = pd.to_datetime(clin_res_mz['Дата создания РКИ'], format='%d.%m.%Y')
print(clin_res_mz.shape)

clin_res_mz=clin_res_mz.join(date_5,how='left', lsuffix='1')
clin_res_mz=clin_res_mz.join(date_6,how='left', lsuffix='1')
clin_res_mz=clin_res_mz.join(date_7,how='left', lsuffix='1')
print(clin_res_mz.shape)
clin_res_mz.head()
```
